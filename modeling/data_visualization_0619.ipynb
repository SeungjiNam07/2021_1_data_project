```python
import pandas as pd
import numpy as np

from shapely.geometry import Point, shape, mapping
from shapely.ops import unary_union

import matplotlib.pyplot as plt
import seaborn as sns

#from geoband.API import *
import folium
import json
import geopandas as gpd
from geopandas import gpd

from folium import plugins
import re
from pandas import ExcelFile
```

```python
data = pd.read_csv("/Users/seungji/Desktop/dataanalysis/project/data/data_spatial.csv")
```

```python
data.columns
```

```python
m = folium.Map(
    location=[37.54238169591348,126.99068745821812],
 attr = '서울',
    zoom_start=12
)

for i in range(len(data)):
    folium.Circle(
    location = [data.lat[i],data.lon[i]],
    radius = 80,
    color = "black").add_to(m)
folium.plugins.Fullscreen(position='topright',  
                   title='FULL SCREEN ON', 
                   title_cancel='FULL SCREEN OFF',
                   force_separate_button=True
                  ).add_to(m)
m
```

```python
m = folium.Map(
    location=[37.54238169591348,126.99068745821812],
 attr = '서울',
    zoom_start=12
)

heatMap = plugins.HeatMap(zip(data.lat, data.lon),
                          min_opacity = 2,
                          radius = 10,
                          color = 'red')

m.add_child(heatMap)


folium.plugins.Fullscreen(position = 'topright',
                  title = 'click to Expand',
                  title_cancel = 'Click to Exit',
                  force_separate_button = True).add_to(m)

m
```
